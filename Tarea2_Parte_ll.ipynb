Primero importamos las bibliotecas que vamos a necesitar.

```python
import numpy as np #librería para cálculos numéricos
import math as mt #esta librería la necesitaremos para poder usar e 
import matplotlib.pyplot as plt
```

Ahora implementamos el método RK4 en una función 

```python
def rk4(f, y0, x0, xf, h):
  n = 100 #número de iteraciones
  x = np.linspace(x0, xf, n) #genera n valores en el rango [x0,xf] 
  y = np.zeros(n)
  y[0] = y0 #asigno la condición inicial
  
  for i in range(n-1):
      k1 = h * f(x[i], y[i])
      k2 = h * f(x[i] + h/2, y[i] + k1/2)
      k3 = h * f(x[i] + h/2, y[i] + k2/2)
      k4 = h * f(x[i] + h, y[i] + k3)
      
      y[i+1] = y[i] + (k1 + 2*k2 + 2*k3 + k4)/ 6
  
 
  return x, y
```

Ahora definimos la función que nos va a definir la ecuación diferencial.

```python
def f(x, y):
    return 2*(1-y)-mt.exp(-4*x)
```

Definimos los parámetros iniciales dados

```python
y0 = 1  # Valor inicial de y
x0 = 0  # Valor inicial de x
xf = 2  # Valor final de x
h = x[1]-x[0]  # Tamaño del paso

x, y = rk4(f, y0, x0, xf, h) #resolvemos, pasando los parámetros que necesita la función, con los valores iniciales ya establecidos.
```

Graficamos las soluciones de la EDO

```python
plt.plot(x, y, 'c')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solución de la EDO utilizando RK4')
plt.show()
```
